## Vemos el numero de combinaciones

**1) Filtramos por las drugs que se encuentren tanto en DRUGCOMB como en Cmaps**

```python
# Obtener los cell_id √∫nicos del primer DataFrame
cp_inchi = pert_info[pert_info["pert_type"] == "trt_cp"]["inchi_key"].unique()


# Filtrar el segundo DataFrame con los nombres que est√°n en cell_info_ids
filtered_drugs = drugs[drugs["inchikey"].isin(cp_inchi)]

# Seleccionar solo las columnas que te interesan
selected_columns_drugs = filtered_drugs[['dname', 'id', 'chembl_id', 'inchikey', 'smiles','molecular_formula','drugbank_id','kegg_id']]

# Mostrar el resultado
print(selected_columns_drugs.shape[0])
selected_columns_drugs.head(2) 
```

**2) Filtramos por las cell lines que se encuentren tanto en DRUGCOMB como en Cmaps**

```python
# Obtener los cell_id √∫nicos del primer DataFrame
cell_info_ids = cell_info[cell_info["cell_type"] == "cell line"]["cell_id"].unique()

# Filtrar el segundo DataFrame con los nombres que est√°n en cell_info_ids
filtered_cell_line = cell_line[cell_line["name"].isin(cell_info_ids)]

# Seleccionar solo las columnas que te interesan
selected_columns_cell = filtered_cell_line[["name", "disease_id", "id", "Tumour_Location", "cellosaurus_tumor_location"]]

# Mostrar el resultado
print(selected_columns_cell.shape[0])
selected_columns_drugs.head(2) 
```

**3) Vemos las combinaciones que hay con estas drogas y cell lines**

```python
# Condici√≥n 1: cell_line_id est√° en selected_columns_cell["id"]

cell_line_filter = summary["cell_line_id"].isin(selected_columns_cell["id"])

# Condici√≥n 2: tanto drug_row_id como drug_col_id est√°n en selected_columns_drugs["id"]
valid_drug_ids = selected_columns_drugs["id"]
drug_row_filter = summary["drug_row_id"].isin(valid_drug_ids)
drug_col_filter = summary["drug_col_id"].isin(valid_drug_ids)

# Aplicar ambas condiciones
filtered_summary = summary[cell_line_filter & drug_row_filter & drug_col_filter]

# Mostrar el resultado
filtered_summary
```

Paso 1 : Merge con selected_columns_drugs (para drug_row_id y drug_col_id)

luego le metes el resto de columnas es para ver que se queda igual


```python
# Merge para drug_row_id
filtered_summary = filtered_summary.merge(
    selected_columns_drugs[["id", "dname", "inchikey"]],
    left_on="drug_row_id",
    right_on="id",
    how="left"
).rename(columns={"dname": "dname_row", "inchikey": "inchikey_row"}).drop(columns=["id"])

# Merge para drug_col_id
filtered_summary = filtered_summary.merge(
    selected_columns_drugs[["id", "dname", "inchikey"]],
    left_on="drug_col_id",
    right_on="id",
    how="left"
).rename(columns={"dname": "dname_col", "inchikey": "inchikey_col"}).drop(columns=["id"])
```

Paso 2: Merge con selected_columns_cell (para cell_line_id)

```python
filtered_summary = filtered_summary.merge(
    selected_columns_cell[["id", "name", "disease_id", "cellosaurus_tumor_location"]],
    left_on="cell_line_id",
    right_on="id",
    how="left"
).rename(columns={
    "name": "cell_line_name",
    "disease_id": "cell_disease_id",
    "cellosaurus_tumor_location": "tumor_location"
}).drop(columns=["id"])
```

```python
print(filtered_summary.shape[0])
filtered_summary.head(2) ## este es el filtered summary con todas las condiciones que queremos 
```

**4) Ahora vemos las cell lines que estan en las combinaciones**

```python
unique_cell_line_ids = filtered_summary["cell_line_id"].unique()
cell_lines_used = selected_columns_cell[selected_columns_cell["id"].isin(unique_cell_line_ids)]
print(cell_lines_used.shape[0])
cell_lines_used
```

**5) Ahora vemos las drugs que estan en las combinaciones**

```python
# Unir todos los IDs de drogas en una sola Serie
all_drug_ids = pd.concat([
    filtered_summary["drug_row_id"],
    filtered_summary["drug_col_id"]
]).unique()
```

```python
unique_drugs_used = selected_columns_drugs[selected_columns_drugs["id"].isin(all_drug_ids)]
unique_drugs_used = unique_drugs_used.sort_values("dname").reset_index(drop=True)
```
