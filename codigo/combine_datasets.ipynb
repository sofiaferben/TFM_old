**2) Load CMAPS LINCS 1000 info**

```python
vae_output=pd.read_csv(r"C:\Users\K6502\OneDrive\Desktop\TFM\datos\VAE\mu_embeddings_transposed.csv")
```

```python
print(vae_output.shape)
vae_output.head(5)
```

```python
import pandas as pd

# Primero ponemos en formato más fácil de buscar el vae_output
# Quitamos 'Unnamed: 0' si está y ponemos 'rid' como índice (opcional)
vae_output = vae_output.drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
vae_output
```

```python
# Transponemos para facilitar el acceso
vae_transposed = vae_output.T
vae_transposed.index.name = 'drug_cell'
```

```python
vae_transposed
```

```python
def get_latents(drug_name, cell_line):
    row_name = f"{drug_name}_{cell_line}"
    if row_name in vae_transposed.index:
        return vae_transposed.loc[row_name].values[:5]
    else:
        return [float('nan')] * 5
```

```python
# Creamos las listas para las nuevas columnas
latent_drug1 = []
latent_drug2 = []

# Iteramos sobre drugcomb_selected_columns
for idx, row in drugcomb_selected_columns.iterrows():
    drug1 = row['cmaps_name_drug1']
    drug2 = row['cmaps_name_drug2']
    cell_line = row['cell_line_name']
    
    # Buscamos latentes
    latents1 = get_latents(drug1, cell_line)
    latents2 = get_latents(drug2, cell_line)
    
    latent_drug1.append(latents1)
    latent_drug2.append(latents2)
```

```python
# Ahora agregamos estas columnas nuevas
latent_drug1_df = pd.DataFrame(latent_drug1, columns=[f"latent_{i}_drug1" for i in range(5)])
latent_drug2_df = pd.DataFrame(latent_drug2, columns=[f"latent_{i}_drug2" for i in range(5)])

# Concatenamos todo
final_df = pd.concat([drugcomb_selected_columns.reset_index(drop=True), latent_drug1_df, latent_drug2_df], axis=1)
```

```python
import pandas as pd

# Definir las columnas clave para hacer el join
join_keys = ['dname_drug1', 'dname_drug2', 'cell_line_name']

# Hacer el merge (inner join)
merged_df = final_df.merge(
    network_medicine,
    on=join_keys,
    how='inner'  # Solo filas donde las claves estén en ambos
)
```
