# COMIBINACIONES DE FÁRMACOS


lo he sacado de la página web https://www.fda.gov/drugs/drug-approvals-and-databases/drugsfda-data-files#download 

```python
import pandas as pd

# Define the file path
file_path = r"C:\Users\Usuario\Desktop\TFM\datos\datfda20250218\Products.txt"


# Read the file and handle inconsistent columns
with open(file_path, "r", encoding="utf-8") as file:
    lines = file.readlines()

# Extract header and determine the expected number of columns
header = lines[0].strip().split("\t")
num_columns = len(header)

# Process each line, ensuring consistent column count
cleaned_lines = [line.strip().split("\t") for line in lines[1:] if len(line.strip().split("\t")) == num_columns]

# Convert the cleaned data into a DataFrame
df = pd.DataFrame(cleaned_lines, columns=header)
```

```python
df.head()
```

```python
# Filter the DataFrame for combination products (DrugName contains "AND" or "||")
df_filtered = df[df["DrugName"].str.contains(r"\bAND\b|\|\|", case=False, na=False, regex=True)]
```

```python
import pandas as pd
```

```python
df_applications = pd.read_csv(r"C:\Users\Usuario\Desktop\TFM\datos\datfda20250218\Applications.txt", delimiter="\t", encoding="ISO-8859-1", on_bad_lines='skip')
df_application_docs = pd.read_csv(r"C:\Users\Usuario\Desktop\TFM\datos\datfda20250218\ApplicationDocs.txt", delimiter="\t", encoding="ISO-8859-1", on_bad_lines='skip')
```

```python
# Convertir la columna "ApplNo" a un tipo de dato consistente en todos los DataFrames
df_filtered = df_filtered.copy()  # Crear una copia explícita antes de modificar
df_filtered.loc[:, "ApplNo"] = df_filtered["ApplNo"].astype(str)
df_applications["ApplNo"] = df_applications["ApplNo"].astype(str)
df_application_docs["ApplNo"] = df_application_docs["ApplNo"].astype(str)

# Realizar los joins entre las tablas usando "ApplNo" como clave
df_merged = df_filtered.merge(df_applications, on="ApplNo", how="left") \
    .merge(df_application_docs, on="ApplNo", how="left")

# Seleccionar columnas relevantes
df_final = df_merged[["DrugName", "ActiveIngredient", "ApplicationDocsTitle", "ApplicationDocsURL"]]
```

```python
df_final
```

```python
# Contar cuántos valores son NaN en la columna ApplicationDocsURL
nan_count = df_final["ApplicationDocsURL"].isna().sum()

# Contar cuántos valores tienen una URL (no son NaN)
url_count = df_final["ApplicationDocsURL"].notna().sum()
print("Numero de combinaciones de drugs sin url: ", str(nan_count),)
print("Numero de combinaciones de drugs  con url: ", str(url_count),)
```
